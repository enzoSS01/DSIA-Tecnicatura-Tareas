```
#Ejercicio 2 – Dataset simple con Pandas
import pandas as pd
datos = {
    'estudiante': ['Ana', 'Juan', 'Carla', 'Pedro', 'Lucia'],
    'nota': [8, 6, 9, 5, 7]
}
df = pd.DataFrame(datos)
print(df)

promedio = df['nota'].mean()
nota_maxima = df['nota'].max()
nota_minima = df['nota'].min()

print(f"Promedio de las notas: {promedio}")
print(f"Nota maxima: {nota_maxima}")
print(f"Nota minima: {nota_minima}")
```

 ¿Qué librería utilizarías para cada una de las siguientes tareas?

pandas → porque permite leer archivos CSV y manipular datasets tabulares\.

seaborn → esta diseñada específicamente para crear gráficos estadísticos y visualizar distribuciones de datos de forma sencilla\.

scikit\-learn → es la librería estándar de Python que contiene los algoritmos y herramientas necesarias para entrenar modelos predictivos\.

numpy → porque proporciona estructuras de datos \(arrays\) altamente eficientes para resolver cálculos numéricos y álgebra lineal a gran velocidad\.
