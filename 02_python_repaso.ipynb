```
"""Ejercicio 1 – Variables y operaciones"""
a = 5
b = 3

print ("El resultado de la suma es:", a + b)
print ("El resultado de la resta es:", b - a)
print ("El resultado de la multiplicacion es:", b * b)
print ("El resultado de la division es:", a / a)
```

```
"""Ejercicio 2 – Condicionales"""
edad = int (input("Ingrese su edad: "))
if edad >= 18:
    print("usted es mayor de edad")
else:
    print("usted es menor de edad")
```

```
"""Ejercicio 3 – Listas y bucles"""
numeros = [10, 25, 30, 45, 50] 
suma_total = 0                 

for num in numeros:
    print(f"{num}")
    suma_total += num  

cantidad_elementos = len(numeros)
promedio = suma_total / cantidad_elementos

print(f"Suma total:{suma_total}")
print(f"Promedio:{promedio}")
```
